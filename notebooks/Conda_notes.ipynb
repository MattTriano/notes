## Reverting in Conda

I accidentally upgraded my base conda environment, rather than the conda env I wanted to upgrade. You can roll back an install via

`!conda list --revisions`

like below

````python
!conda list --revisions
````

Identify the revision number you want to revert to, then enter the command (but replace the `2` with the revision number you want to revert to

`conda install --revision 2`

# How to install packages into a jupyter kernel from jupyter notebook or lab

per https://jakevdp.github.io/blog/2017/12/05/installing-python-packages-from-jupyter/

You need to indicate the prefix to install packages (eg `psycopg2`) into. 

```bash
!conda install --channel conda-forge --yes --prefix {sys.prefix} psycopg2
```
